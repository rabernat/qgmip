```python
import numpy as np
from matplotlib import pyplot as plt
from pyqg import qg_model
from IPython.display import display, clear_output
%matplotlib inline
```

```python
m = qg_model.QGModel()
m = qg_model.QGModel(nx=256, L=2*np.pi, beta=0., rek=0., dt=2e-3)
q1 = -np.cos(m.x) + np.cos(m.x + m.y) # + 1e-3 * np.random.rand(m.ny, m.nx)
m.set_q1q2(q1, q1)
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.axes((0,0,1,1))
i = 0
for n in m.run_with_snapshots(tsnapint=100*m.dt):
    ax.cla()
    im = ax.imshow(m.q1, origin='bottom', cmap='bone')
    im.set_clim([-2,2])
    ax.set_axis_off()
    fig.savefig('animations/barotropic_initial_value/q1_%05d.png' % i)
    i += 1
```

```python
ax = plt.subplot(111)
ax.cla()
im=ax.imshow(-np.cos(m.x) + np.cos(m.x + m.y) + 1e-3 * np.random.rand(m.ny, m.nx),
          origin='bottom', cmap='bone')
ax.set_xticks([])
ax.set_yticks([])
plt.colorbar(im)
```
