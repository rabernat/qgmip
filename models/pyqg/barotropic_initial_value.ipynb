```python
import numpy as np
from matplotlib import pyplot as plt
from pyqg import qg_model
from IPython.display import display, clear_output
%matplotlib inline
```

```python
m = qg_model.QGModel()
m = qg_model.QGModel(nx=256, L=2*np.pi, beta=1.0, rek=0., dt=1./(2.**10), U1=0.0, U2=0.0, delta=1, tmax=20)
q1 = -3*np.cos(m.x) + 3*np.cos(m.x + m.y) #Notice that our domain is x,y=(0,2*pi), hence the minus sign in the first term
m.set_q1q2(q1, q1)
```

```python
fig = plt.figure(figsize=(12,12))
ax = plt.axes((0,0,1,1))
i = 0
for n in m.run_with_snapshots(tsnapint=100*m.dt):
    ax.cla()
    im = ax.imshow(m.q1, origin='bottom', cmap='bone')
    im.set_clim([-6,6])
    ax.set_axis_off()
    clear_output(wait=True)
    display(fig)
    fig.savefig('animations/barotropic_initial_value/256x256/q1_%05d.png' % i)
    i += 1
```
