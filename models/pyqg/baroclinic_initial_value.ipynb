```python
import numpy as np
import time
from matplotlib import pyplot as plt
from pyqg import qg_model
from IPython.display import display, clear_output
%matplotlib inline
```

```python
m = qg_model.QGModel()
m = qg_model.QGModel(nx=256, L=4*np.pi, beta=1.0, rd=0.5, rek=0., dt=1./(2.**9), U1=0.0, U2=0.0, delta=0.2, tmax=100)
#q1 = 3*np.cos(m.x) + 3*np.cos(m.x + m.y) 
q1 = 3*np.cos(m.x) + 3*np.cos(m.x + m.y) + 0.1*np.random.rand(m.ny,m.nx)-0.15 #Perturbed initial conditions
q2 = -3*np.cos(m.x) + 3*np.cos(m.x - m.y)
m.set_q1q2(q1, q2)
```

```python
#fig = plt.figure(figsize=(12,12))
#ax = plt.axes((0,0,1,1))
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(13,6))
i = 0
start_time = time.time()
for n in m.run_with_snapshots(tsnapint=100*m.dt):
    ax1.cla()
    im1 = ax1.imshow(m.q1+m.beta1*m.y, origin='bottom', cmap='bone')
    im1.set_clim([-6,18])
    ax1.set_aspect('equal')
    ax1.set_axis_off()
    ax2.cla()
    im2 = ax2.imshow(m.q2+m.beta2*m.y, origin='bottom', cmap='bone')
    im2.set_clim([-6,18])
    ax2.set_aspect('equal')
    ax2.set_axis_off()
    clear_output(wait=True)
    display(fig)
    #fig.savefig('animations/baroclinic_initial_value/256x256_pert/q1_%05d.png' % i)
    i += 1
end_time=time.time()
print end_time - start_time
# This took about 37 min on my macbook pro with a 2.66GHz dual-core I7 processor (as always I did also have a bunch of other stuff open though)
```
